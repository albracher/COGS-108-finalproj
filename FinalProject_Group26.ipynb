# Data Wrangling

We start our data wrangling by importing Pandas, NumPy, Pyplot, and seaborn. 

```python
## import pandas, numpy, pyplot, seaborn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Next, we import our SCF and public transportation datasets (SDG indicator).

```python
# read in our CSV file for the SCF
SCF_median = pd.read_csv('https://raw.githubusercontent.com/albracher/cogs_108_final_project_csvs/main/interactive_bulletin_charts_inccat_median.csv'
)

# read in our CSV file for those with access to public transportation 
public_transportation = pd.read_csv('https://raw.githubusercontent.com/charbuurst/COGS108_Repo/main/11_2_1_Percentage_Access_to_Public_Transport.csv')
```

# Data Cleaning

```python
# check top rows of public transportation
public_transportation.head() 
# check columns of public transportation
public_transportation.columns
```

```python

# look at columns within SCF
SCF_median.columns

# drop na rows
SCF_median = SCF_median.dropna()
#drop duplicates
SCF_median = SCF_median.drop_duplicates()

#python list of columns to delete
columns_to_drop = ["Installment_Loans", "Mortgages_Home_Equity_Loans", "Vehicle_Installment_Loans", "Home_Equity_Lines_of_Credit",
                  "Education_Installment_Loans", "Other_Installment_Loans", "Other_Non_Financial_Assets", "Nonfinancial_Assets",]

#use function to drop those columns
SCF_median = SCF_median.drop(columns=columns_to_drop)
```

```python

# check for missing values 
print(public_transportation.isnull().sum())

# check for duplicates 
print(public_transportation.duplicated().sum())
```

```python

# drop leftmost and rightmost columns 
columns_to_drop_transport = ['Indicator', 'Estimate_source']
public_transportation = public_transportation.drop(columns = columns_to_drop_transport)
public_transportation.head()
```

```python
# drop rows with missing values
public_transportation.dropna(subset = ['Country', 'Region', 'Cities'])

# only want the United States 
US_public_transportation = public_transportation[public_transportation['Country'] == 'United States']
US_public_transportation = US_public_transportation.reset_index()
US_public_transportation
```

# Data Visualization/Basic Data Analysis

```python
# line graph on primary residence vs time
# Primary_Residence = value of their home
sns.lineplot(data = SCF_median, x = 'year', y = 'Primary_Residence').set(title='Median Value of Primary Residences in the U.S. Over Time')
```

"Primary Residences" in this case refers to the value of one's home; both rented and owned homes fall under "primary residences". As can be seen by the graph, there was a spike in value in around 2002 to 2004, followed by a decrease from 2005 to 2012. By 2013 onwards, the median value of primary residences steadily increased again.

```python
# line graph on primary residence vs time, grouped by income
sns.lineplot(data = SCF_median, x = 'year', y = 'Primary_Residence', hue = 'Category').set(title='Median Value of Primary Residences in the U.S. Over Time, Grouped By Income Percentile')
```

When looking at the median value of primary residences over time, a similar trend can be observed. However, when grouping by different income profiles, we notice that the 90-100th percentile of income's "primary residence" value is far higher than the 80-89.9th percentile and under.

```python
# line graph on owned vehicles vs time
# same thing on owned vehicles, I'd assume that they're looking at the value of all owned vehicles
sns.lineplot(data = SCF_median, x = 'year', y = 'Owned_Vehicles').set(title='Median Value of Owned Vehicles  in the U.S. Over Time')
```

Observing the median value of owned vehicles in the U.S. over time, we notice similar trends to the median value of primary residences. There is an steady and sharp increase in owned vehicle value from 1999 to around 2005, and sharp decrease from 2005 to around 2008, then relative steadiness from 2009 onwards.

```python
# line graph on owned vehicles vs time, group by percentile of income
sns.lineplot(data = SCF_median, x = 'year', y = 'Owned_Vehicles', hue = 'Category').set(title='Median Value of Owned Vehicles in the U.S. Over Time, Grouped by Income Percentile')
```

After grouping the above data by income percentile, we, again, notice similar trends. However, among those whose income resides in the 80-100th percentile, from 2005, onwards, the median value of their owned vehicles is more volatile. There is a trend of a noticable dip around 2008, followed by a small increase in median value, followed by a large dip around 2013, and then another increase/decrease in median value.

```python
US_public_transportation = US_public_transportation.sort_values('Percentage_Value', ascending=False)

plt.figure(figsize=(10,5))
plt.bar(US_public_transportation['Cities'], US_public_transportation['Percentage_Value'])
plt.xlabel('Cities')
plt.ylabel('Percentage Value')
plt.title('Percentage of Those With Access to Public Transportation by City')
plt.xticks(rotation=90)
plt.show()
```

People who live in cities with greater accessibility to public transportation tend to rely less on personal vehicles since public transportation is the cheaper alternative. The cities with a higher percentage of access to public transporation (Gainesville, Manchester, Savanna) should be looked into to see if there are any interactions or trends with vehicle spending.
